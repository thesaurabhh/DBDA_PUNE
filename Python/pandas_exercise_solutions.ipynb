Load the lms_members.xlsx file to Python using pandas and perform the following operation:

/* Display name of all the members */
/* Display name of all the members whose membership_status is Permanent*/
/* Display Name of the member , City in which he resides and date of membership expiry*/
/* name of the member whose date_expire is greater than 01-01-2020 */
/* name of the member who resides in PUNE */
/* name of the member who resides in Pune and his membership_status is Permanent*/
/* list all unique cities in the members dataframe*/

```python
import pandas as pd
df= pd.read_excel("LMS.xlsx","lms_members") # read a particular sheet of the excel file named LMS.xlsx
df.head()
```

```python
# /* Display name of all the members */
df["MEMBER_NAME"]
```

```python
#/* Display name of all the members whose membership_status is Permanent*/
df[df['MEMBERSHIP_STATUS']== 'Permanent'][["MEMBER_NAME","MEMBERSHIP_STATUS"]]
```

```python
#/* Display Name of the member , City in which he resides and date of membership expiry*/
df[["MEMBER_NAME","CITY","DATE_EXPIRE"]]
```

```python
#/* name of the member whose date_expire is greater than 01-01-2020 */
df[pd.to_datetime(df["DATE_EXPIRE"],format="%Y-%m-%d") > pd.to_datetime('01-01-2020',format="%d-%m-%Y")][["MEMBER_NAME","DATE_EXPIRE"]]
```

```python
#/* name of the member who resides in PUNE */
#df[df["CITY"].str.lower() == "pune" ][["MEMBER_NAME","CITY"]]
df[df["CITY"].isin(["Pune","PUNE"]) ][["MEMBER_NAME","CITY"]]
```

```python
#/* name of the member who resides in Pune and his membership_status is Permanent
#df[(df.points > 20) & (df.assists == 9)]
df[ (df["CITY"].isin(["Pune","PUNE"])) & (df["MEMBERSHIP_STATUS"] == "Permanent" ) ][["MEMBER_NAME","MEMBERSHIP_STATUS"]]
```

```python
#/* list all unique cities in the members dataframe*/
df["CITY"].drop_duplicates()
```
